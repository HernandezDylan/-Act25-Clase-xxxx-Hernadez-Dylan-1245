# Crear la clase Evento, con 3 atributos (nombre, fecha, lugar), 2 funciones leerdatos(), mostrardatos() y 2 objetos (evento1, evento2).



<img src="https://raw.githubusercontent.com/HernandezDylan/-Act25-Clase-xxxx-Hernadez-Dylan-1245/refs/heads/main/infoclassxxx.png.png" witdh="500" height="400" alt="Texto descriptivo">




```python
print("Clase Evento")

class Evento:
    def __init__(self):
        self.nombre = ""
        self.fecha = ""
        self.lugar = ""

    def leerdatos(self):
        self.nombre = input("Nombre: ")
        self.fecha = input("Fecha: ")
        self.lugar = input("Lugar: ")

    def mostrardatos(self):
        print(self.nombre, self.fecha, self.lugar)

evento1 = Evento()
evento2 = Evento()

while True:
    op = input("\n1.Ev1 2.Mostrar1 3.Ev2 4.Mostrar2 5.Salir: ")

    if op == "1":
        evento1.leerdatos()

    elif op == "2":
        evento1.mostrardatos()

    elif op == "3":
        evento2.leerdatos()

    elif op == "4":
        evento2.mostrardatos()

    elif op == "5":
        print("Programa realizado por Hernandez Dylan 1245")
        break

    else:
        print("Error")
```
